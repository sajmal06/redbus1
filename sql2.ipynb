```python
import pandas as pd
from openpyxl import Workbook
from openpyxl import load_workbook
import re as re

wb = load_workbook('Routes.xlsx')
sheet1 = wb['Sheet1']
df1 = pd.DataFrame(sheet1.values)
df1.columns = df1.iloc[0]
df1 = df1[1:]
# df1 = df1.drop(columns=['NaN'],inplace=True)
df1.iloc[0]['Route']
```

```python
#created table for separate routes and related links
mycursor.execute("CREATE TABLE REDBUS.ROUTES (Route varchar(255), Link varchar(255))")
```

```python
#inserted the route and route link file into sql under table name Routes
for index, row in df1.iterrows():
    rou = row['Route']
    rou2 = row['Link']
    query = "INSERT INTO REDBUS.ROUTES (Route, Link) VALUES (%s, %s)"
    mycursor.execute(query, (rou, rou2))
mydb.commit()

```

```python
query = "UPDATE REDBUS.ROUTES SET Route = REPLACE(Route,' ','_')"
mycursor.execute(query)
mydb.commit()
out = mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out))
```
